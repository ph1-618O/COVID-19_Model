```python
#Import the libraries required
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Importing the excel data 
dataset = pd.read_csv("no_nans3.csv")
dataset
# dataset.dtypes
# dataset['total_population'] = dataset['total_population'].astype(int) 

```

```python
dataset[['population','cases','deaths']]
```

```python
def assess_NA(data):
    null_sum=data.isnull().sum()
    total = null_sum.sort_values(ascending=False)
    percent = ( ((null_sum / len(data.index))*100).round(2) ).sort_values(ascending=False)
    df_NA = pd.concat([total, percent], axis=1, keys=['Number of NA', 'Percent NA'])
    df_NA = df_NA[ (df_NA.T != 0).any() ]
    
    return df_NA

```

```python
df_NA= assess_NA(dataset)
df_NA
```

```python
X = dataset[["cases"]]
y = dataset["deaths"].values.reshape(-1, 1)
print(X.shape, y.shape)
```
